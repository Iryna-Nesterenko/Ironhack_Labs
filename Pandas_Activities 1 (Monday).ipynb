```python
file1 = pd.read_csv('/Users/irenewalken/Documents/GitHub/IH_RH_DA_FT_AUG_2022/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/file1.csv')
file2 = pd.read_csv('/Users/irenewalken/Documents/GitHub/IH_RH_DA_FT_AUG_2022/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/file2.csv')
file3 = pd.read_csv('/Users/irenewalken/Documents/GitHub/IH_RH_DA_FT_AUG_2022/Class_Materials/Case_Studies/Customer_Analysis_Case_Study/Data/file3.csv')
```

```python
file2.rename(columns = {'ST':'State', 'GENDER':'Gender'}, inplace = True)
file1.rename(columns = {'ST':'State','GENDER':'Gender'}, inplace = True)
```

```python
file1 = file1[['State','Gender','Education','Customer Lifetime Value','Income','Monthly Premium Auto','Number of Open Complaints','Policy Type','Vehicle Class','Total Claim Amount']]
file2 = file2[['State','Gender','Education','Customer Lifetime Value','Income','Monthly Premium Auto','Number of Open Complaints','Policy Type','Vehicle Class','Total Claim Amount']]
file3 = file3[['State','Gender','Education','Customer Lifetime Value','Income','Monthly Premium Auto','Number of Open Complaints','Policy Type','Vehicle Class','Total Claim Amount']]
```

```python
file2
```

```python
def customer_data():
   
    return pd.concat([file1,file2,file3], axis=0)
```

```python
df = customer_data()
```

```python
df
```

```python
df.info()
```

```python
df['Customer Lifetime Value']= df['Customer Lifetime Value'].apply(lambda x: x.strip('%') if isinstance(x,str) else x)#remove the percentage from the customer lifetime value
```

```python
df['Customer Lifetime Value'] =  pd.to_numeric(df['Customer Lifetime Value'], errors='coerce') #transform it to float
```

```python
df['Customer Lifetime Value']= df['Customer Lifetime Value'].apply(lambda x: float(round(x,0)) if isinstance(x,(int,float)) else x)#transform to int and nan stay float
```

```python
df["Number of Open Complaints"].unique()
```

```python
df["Number of Open Complaints"] = df["Number of Open Complaints"].str.split('/').str[1] #clean the number of open complaints and extract the middle number which is changing between records. pay attention that the number of open complaints is a categorical feature.
```

```python
df["Number of Open Complaints"].unique()
```

```python
df["State"].value_counts()
```

```python
df["State"]= df["State"].replace(["WA","AZ","Cali"], ["Washington", "Arizona","California"])
```

```python
df["Gender"].value_counts()
```

```python
df["Gender"]=df["Gender"].replace(["Male","female","Femal"],["M","F","F"])
```
